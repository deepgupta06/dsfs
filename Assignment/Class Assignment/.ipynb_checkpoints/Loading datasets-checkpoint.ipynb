### 10. From a  cabin column seperate text and numeric value .

```python
import re
```

```python
df["Cabin"].fillna("NaN", inplace=True)
```

```python
df["Cabin_list"] = df["Cabin"].apply(lambda x: re.split(r"(\d+)",x.strip()) if x !="NaN" else "")
df["Cabin_txt"] = ""
df["Cabin_number"] = ""
for i in range(len(df["Cabin_list"])):
    #print(df["Cabin_list"].iloc[i])
    if type(df["Cabin_list"].iloc[i]) == list:
        txt = []
        number = []
        for element in df["Cabin_list"].iloc[i]:
            if str(element).isnumeric():
                number.append(str(element))
            else:
                #print(element)
                if element =="":
                    pass
                else:
                    if len(txt)==0:
                        txt.append(element.strip())
                    else:
                        if txt[0] == element.strip():
                            pass
                        else:
                            txt.append(element.strip())
        df["Cabin_txt"][i] = ",".join(txt)
        df["Cabin_number"][i] = ",".join(number)
```
